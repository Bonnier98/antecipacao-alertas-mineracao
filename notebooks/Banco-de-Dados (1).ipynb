Primeiramente vamos criar e configurar nosso schema que seja nossa base de dados de forma que mantenha um padrao de segurança e integridade dos dados brutos iniciais.

```python
#definindo o catalogo que sera usado
spark.sql("USE CATALOG `estudo-guiado`")
```

```python
# criando schema "vale" de forma isolada, utilizando do 'IF NOT EXISTS'  para garantir que o codigo pode rodar 100 vezes sem quebrar o que já existe
spark.sql("CREATE SCHEMA IF NOT EXISTS vale")
```

```python
# garantir propriedade de isolamento e integridade no nosso ambiente
spark.sql("USE SCHEMA vale")

print("====================")
print("Crowley / Governanca de dados - Status: PRONTO")
print("Catálogo Ativo: estudo-guiado")
print("Schema Ativo: vale")
print("Propriedade de Integridade: Ativadas")
print("====================")
```

Agora com o schema que representara nosso banco de dados criado vamos criar os demais schemas que serao nossas camda da arquitetura Medalhão que iremos trabalhar.

```python
#selecionamos o catalogo correto
spark.sql("USE CATALOG `estudo-guiado`")
```

```python
#criamos as camadas da arquitetura medalhão
spark.sql("CREATE SCHEMA IF NOT EXISTS vale_bronze")
spark.sql("CREATE SCHEMA IF NOT EXISTS vale_silver")
spark.sql("CREATE SCHEMA IF NOT EXISTS vale_gold")

print("====================")
print("Crowley / Governanca de dados - Status: PRONTO")
print("Catálogo Ativo: estudo-guiado")
print("Schema Ativo: vale_bronze, vale_silver, vale_gold")
print("Propriedade de Integridade: Ativadas")
print("====================")
```
